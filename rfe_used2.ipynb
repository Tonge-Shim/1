```
logreg=LogisticRegression()
selector=RFE(logreg)
selector=selector.fit(x,y)#99등까지의 것을 다시 selector 함수에 대입
```

```
selected_features=DataFrame({'Feature': list(x.columns), 'Ranking': selector.ranking_})
selected_features.sort_values(by='Ranking').head(50)
```

```
newx=selected_features.sort_values(by='Ranking')['Feature']
newx=newx.loc[197:40]#ranking2ndround
newx.tolist()
X=x[newx]

print(X.shape)
print(y.shape)
X
X
```

```
sm = SMOTE(ratio='auto', kind='regular')
X_re, y_re = sm.fit_sample(X,y)
```

```
X_train, X_test, Y_train, Y_test=train_test_split(X_re,y_re,random_state=30)#99feature, 6610samples
x_train, x_test, y_train, y_test=train_test_split(X,y, random_state=30)#99feature, 1924 samples
#print(X_train.shape)
print(x_test.shape)
print(x_train.shape)
#print(y_train.shape)
#print(y_test.shape)
```

```
logreg=LogisticRegression()
y_pred=logreg.fit(X_train,Y_train).predict(x_test)
print('The Accuracy of the model is {:.1f}%'.format(accuracy_score(y_test,y_pred)*100))
```
